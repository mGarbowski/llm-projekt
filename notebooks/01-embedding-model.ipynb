## Modele są ograniczone co do liczby tokenów

* Dłuższy chyba(?) jest przycinany
* Maksymalna długość dotyczy liczby tokenów
* Metoda tokenizacji jest zależna od modelu

```python
model.max_seq_length
```

```python
model.tokenizer
```

```python
text = chunks[1]
len(text)
```

```python
model.tokenize(text)["input_ids"].shape[0]
```

```python
different_len_embeddings = [
    model.encode(text[:i])
    for i in range(1, len(text))
]

for i, emb in enumerate(different_len_embeddings[1:], start=1):
    prev_embedding = different_len_embeddings[i-1]
    is_the_same = np.allclose(emb, prev_embedding)
    number_of_tokens = model.tokenize(text[:i])["input_ids"].shape[0]
    if is_the_same:
        print(f"takie samo dla długości {i} i {i+1} (liczba tokenow to {number_of_tokens})")
```
